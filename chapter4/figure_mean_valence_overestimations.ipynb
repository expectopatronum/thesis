```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams.update({'font.size': 12})
```

```python
categories = ["DEAM", "PMEmo+DEAM"]
mean_valence_overestimation = {
    "hiphop": [0.395, 0.33],
    "other": [0.20, 0.20]
}
# error_bars = {
#     "hiphop": [0.02, 0.02],  
#     "other": [0.05, 0.05]  
# }

error_bars = {
    "hiphop": [0.011, 0.016],  
    "other": [0.028, 0.0215]  
}

# Bar width
bar_width = 0.4
x = np.arange(len(categories))

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6.4, 3.8), dpi=300)

# Plot bars with error bars
ax.bar(x - bar_width/2, mean_valence_overestimation["hiphop"], bar_width, yerr=error_bars["hiphop"],
       label="hiphop", color="tab:blue", capsize=5)
ax.bar(x + bar_width/2, mean_valence_overestimation["other"], bar_width, yerr=error_bars["other"],
       label="other", color="tab:orange", capsize=5)

# Labels and formatting
ax.set_xlabel("Training Dataset")
ax.set_ylabel("Mean Valence Overestimation")
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend(loc='upper right', ncols=2)

# Add horizontal grid lines
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
